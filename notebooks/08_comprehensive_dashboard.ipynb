```python
# Calculate execution rates and other metrics
if "budget_execution" in data:
    budget_df = data["budget_execution"].copy()
    budget_df["execution_rate"] = (budget_df["execution"] / budget_df["budget"]) * 100
    budget_df["underspent"] = budget_df["budget"] - budget_df["execution"]
    
    print("Budget execution data with calculated metrics:")
    print(budget_df[['year', 'sector', 'budget', 'execution', 'execution_rate', 'underspent']].head(10))
```

```python
# Dashboard: Combined view with all key metrics

# Create a comprehensive dashboard with multiple panels
fig = make_subplots(
    rows=3, cols=2,
    subplot_titles=(
        'EjecuciÃ³n Presupuestaria por AÃ±o', 
        'DistribuciÃ³n de Gastos por Sector',
        'Indicadores ProgramÃ¡ticos', 
        'Licitaciones por Mes',
        'Tasa de EjecuciÃ³n vs Promedio', 
        'Presupuesto vs Ejecutado por Sector'
    ),
    specs=[[{"secondary_y": True}, {"type": "pie"}],
           [{"secondary_y": True}, {"secondary_y": True}],
           [{}, {"type": "bar"}]],
    vertical_spacing=0.1
)

# Panel 1: Budget execution by year
if "budget_execution" in data:
    budget_summary = budget_df.groupby('year').agg({
        'budget': 'sum',
        'execution': 'sum'
    }).reset_index()
    budget_summary['execution_rate'] = (budget_summary['execution'] / budget_summary['budget']) * 100
    
    fig.add_trace(
        go.Scatter(x=budget_summary['year'], y=budget_summary['execution_rate'],
                  mode='lines+markers', name='Tasa de EjecuciÃ³n', line=dict(color='blue', width=4)),
        row=1, col=1
    )
    
    # Add horizontal line at 95%
    fig.add_hline(y=95, line_dash="dash", line_color="red", row=1, col=1)
    
    # Add annotation for uniform execution rates
    fig.add_annotation(x=2020.5, y=96, text="Tasa uniforme (>95%)", showarrow=True, arrowhead=1, row=1, col=1)

# Panel 2: Spending distribution by sector
if "budget_execution" in data:
    sector_summary = budget_df.groupby('sector').agg({
        'execution': 'sum'
    }).reset_index()
    
    fig.add_trace(
        go.Pie(labels=sector_summary['sector'], values=sector_summary['execution'], name="Gastos por Sector"),
        row=1, col=2
    )

# Panel 3: Programmatic indicators
if "indicators" in data:
    ind_df = data["indicators"].copy()
    ind_df['gap'] = ind_df['executed'] - ind_df['planned']
    
    # Group by indicator type
    ind_summary = ind_df.groupby('indicator').agg({
        'planned': 'sum',
        'executed': 'sum',
        'gap': 'sum'
    }).reset_index()
    
    # Add bars for planned vs executed
    fig.add_trace(
        go.Bar(x=ind_summary['indicator'], y=ind_summary['planned'],
               name='Planificado', marker_color='lightblue'),
        row=2, col=1
    )
    
    fig.add_trace(
        go.Bar(x=ind_summary['indicator'], y=ind_summary['executed'],
               name='Ejecutado', marker_color='pink'),
        row=2, col=1
    )

# Panel 4: Procurement timeline
if "procurement" in data:
    procurement_df = data["procurement"].copy()
    procurement_df['month'] = procurement_df['date'].dt.to_period('M')
    
    monthly_procurement = procurement_df.groupby('month').agg({
        'value': 'sum',
        'id': 'count'
    }).reset_index()
    monthly_procurement['month_str'] = monthly_procurement['month'].astype(str)
    
    fig.add_trace(
        go.Scatter(x=monthly_procurement['month_str'], y=monthly_procurement['value'],
                  mode='lines+markers', name='Valor Total', line=dict(color='green', width=4)),
        row=2, col=2
    )
    
    # Add annotation for November concentration
    nov_data = monthly_procurement[monthly_procurement['month_str'] == '2023-11']
    if len(nov_data) > 0:
        fig.add_annotation(x=nov_data['month_str'].iloc[0], y=nov_data['value'].iloc[0], 
                          text="Pico de Noviembre", showarrow=True, arrowhead=1, row=2, col=2)

# Panel 5: Execution rate vs average
if "budget_execution" in data:
    avg_execution_rate = budget_summary['execution_rate'].mean()
    
    fig.add_trace(
        go.Bar(x=budget_summary['year'], y=budget_summary['execution_rate'],
               name='Tasa de EjecuciÃ³n', marker_color='lightcoral'),
        row=3, col=1
    )
    
    # Add horizontal line for average
    fig.add_hline(y=avg_execution_rate, line_dash="dash", line_color="blue", row=3, col=1,
                  annotation_text=f"Promedio: {avg_execution_rate:.2f}%")

# Panel 6: Budget vs execution by sector
if "budget_execution" in data:
    sector_budget_exec = budget_df.groupby('sector').agg({
        'budget': 'sum',
        'execution': 'sum'
    }).reset_index()
    
    fig.add_trace(
        go.Bar(x=sector_budget_exec['sector'], y=sector_budget_exec['budget'],
               name='Presupuestado', marker_color='lightblue'),
        row=3, col=2
    )
    
    fig.add_trace(
        go.Bar(x=sector_budget_exec['sector'], y=sector_budget_exec['execution'],
               name='Ejecutado', marker_color='lightgreen'),
        row=3, col=2
    )

# Update layout
fig.update_layout(
    title="Dashboard Integral de Transparencia - Carmen de Areco",
    height=1200,
    showlegend=True
)

fig.show()
fig.write_html("../public/charts/comprehensive_transparency_dashboard.html")
```

```python
# Create summary metrics
print("\n\U0001f4ca MÃ©tricas Clave del Dashboard:")
print("=" * 50)

# Overall execution rate
if "budget_execution" in data:
    total_budget = budget_df['budget'].sum()
    total_execution = budget_df['execution'].sum()
    overall_execution_rate = (total_execution / total_budget) * 100
    
    print(f"\nðŸ“Š Tasa de EjecuciÃ³n General: {overall_execution_rate:.2f}%")
    print(f"   Presupuesto Total: ${total_budget:,.2f}")
    print(f"   Ejecutado Total: ${total_execution:,.2f}")
    
    # Uniform execution rates analysis
    sector_execution_rates = budget_df.groupby('sector')['execution_rate'].mean()
    std_execution_rate = sector_execution_rates.std()
    
    print(f"\nðŸ“ˆ DesviaciÃ³n EstÃ¡ndar de Tasa de EjecuciÃ³n por Sector: {std_execution_rate:.2f}%")
    if std_execution_rate < 5:
        print("   âš ï¸  Tasa de ejecuciÃ³n muy uniforme - posiblemente contabilidad de compromisos")

# Gender spending analysis
print(f"\n\U0001f6a9 AnÃ¡lisis de Gasto con Perspectiva de GÃ©nero:")
print("   Requiere datos adicionales sobre gastos con perspectiva de gÃ©nero")

# Procurement analysis
if "procurement" in data:
    procurement_df = data["procurement"].copy()
    total_procurement_value = procurement_df['value'].sum()
    nov_procurement = procurement_df[procurement_df['date'].dt.month == 11]
    nov_value = nov_procurement['value'].sum()
    
    print(f"\n\U0001f6a9 AnÃ¡lisis de Licitaciones:")
    print(f"   Valor total de licitaciones: ${total_procurement_value:,.2f}")
    print(f"   Valor en noviembre: ${nov_value:,.2f} ({(nov_value/total_procurement_value)*100:.2f}% del total)")
    print(f"   âš ï¸  {len(nov_procurement)} licitaciones concentradas en 15 dÃ­as de noviembre")

# Programmatic indicators analysis
if "indicators" in data:
    ind_df = data["indicators"].copy()
    ind_df['gap'] = ind_df['executed'] - ind_df['planned']
    
    print(f"\n\U0001f4f0 Indicadores ProgramÃ¡ticos:")
    indicator_summary = ind_df.groupby('indicator').agg({
        'planned': 'sum',
        'executed': 'sum',
        'gap': 'sum'
    }).reset_index()
    
    for _, row in indicator_summary.iterrows():
        gap_pct = (row['gap'] / row['planned']) * 100 if row['planned'] > 0 else 0
        print(f"   {row['indicator']}: Planificado={row['planned']}, Ejecutado={row['executed']}, Diferencia={gap_pct:.2f}%")
```

```python
# Create a detailed dashboard with multiple views
print("\n\n\U0001f50d Creando vistas detalladas del dashboard...")

# 1. Sector analysis
if "budget_execution" in data:
    sector_analysis = budget_df.groupby('sector').agg({
        'budget': 'sum',
        'execution': 'sum',
        'execution_rate': 'mean'
    }).reset_index()
    
    sector_analysis['admin_ratio'] = sector_analysis.apply(
        lambda row: row['budget'] / sector_analysis[sector_analysis['sector'] == 'AdministraciÃ³n']['budget'].iloc[0] 
        if 'AdministraciÃ³n' in sector_analysis['sector'].values else 0, axis=1
    )
    
    print("\n\U0001f4c9 AnÃ¡lisis por Sector:")
    print(sector_analysis[['sector', 'budget', 'execution', 'execution_rate', 'admin_ratio']].round(2))

# 2. Yearly trend analysis
if "budget_execution" in data:
    yearly_trend = budget_df.groupby('year').agg({
        'budget': 'sum',
        'execution': 'sum'
    }).reset_index()
    
    yearly_trend['execution_rate'] = (yearly_trend['execution'] / yearly_trend['budget']) * 100
    yearly_trend['budget_growth'] = yearly_trend['budget'].pct_change() * 100
    yearly_trend['execution_growth'] = yearly_trend['execution'].pct_change() * 100
    
    print("\n\U0001f4c8 Tendencia Anual:")
    print(yearly_trend[['year', 'budget', 'execution', 'execution_rate', 'budget_growth', 'execution_growth']].round(2))

# 3. Summary dashboard
print("\n\n\U0001f3af Dashboard Resumen:")
print("=" * 40)

if "budget_execution" in data:
    print(f"ðŸ’° Presupuesto Total: ${total_budget:,.2f}")
    print(f"ðŸ’¸ Ejecutado Total: ${total_execution:,.2f}")
    print(f"ðŸ“Š Tasa de EjecuciÃ³n: {overall_execution_rate:.2f}%")
    
if "procurement" in data:
    print(f"\n\U0001f6a8 Licitaciones: {len(procurement_df)}")
    print(f"   Valor Total: ${total_procurement_value:,.2f}")
    print(f"   Concentradas en Noviembre: {len(nov_procurement)}")
    
if "indicators" in data:
    print(f"\n\U0001f4f0 Indicadores: {len(indicator_summary)} tipos")
    for _, row in indicator_summary.iterrows():
        print(f"   {row['indicator']}: {row['executed']}/{row['planned']} ({(row['executed']/row['planned'])*100:.1f}%)" if row['planned'] > 0 else f"   {row['indicator']}: {row['executed']}")
```
